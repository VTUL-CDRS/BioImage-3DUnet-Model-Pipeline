```python
from skimage import morphology
import tifffile as tif
import numpy as np
import matplotlib.pyplot as plt
from pathlib import Path
```

```python
fig, ax = plt.subplots()
for cc in [300, 500]:
    for thr in [30, 60, 127]:
        file = f'stat_{cc}_{thr}.npz'
        stats = np.load(Path('prediction') / 'control3' / file)
        ax.hist(stats['cnts'], bins=100, range=(200, 1000), histtype='step', alpha=0.8, label=f'{cc}_{thr}')
ax.legend()
ax.set_title('control3 sizes')
fig.show()
```

```python
fig, ax = plt.subplots()
for cc in [300, 500]:
    for thr in [30, 60, 127]:
        file = f'stat_{cc}_{thr}.npz'
        stats = np.load(Path('prediction') / 'control3' / file)
        ax.hist(stats['nn_distances'], bins=100, range=(0, 80), histtype='step', alpha=0.8, label=f'{cc}_{thr}')
ax.legend()
ax.set_title('control3 distances')
fig.show()
```

```python
nums = []
labels = []
for cc in [300, 500]:
    for thr in [30, 60, 127]:
        file = f'stat_{cc}_{thr}.npz'
        stats = np.load(Path('prediction') / 'rheb2' / file)
        nums.append(stats['nn_distances'].shape[0])
        labels.append(f'{cc}_{thr}')

plt.figure()
plt.bar(labels, nums, label=labels)
plt.title('rheb2')
plt.show()
```
